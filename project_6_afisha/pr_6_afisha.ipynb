```python
#рассчитаем среднюю активность пользователей в день, месяц и неделю
mau = visits.groupby(['session_year', 'session_month']).agg({'uid': 'nunique'})
wau = visits.groupby(['session_year', 'session_week']).agg({'uid': 'nunique'})
dau = visits.groupby('session_date').agg({'uid': 'nunique'})

mau_total = mau.mean()
wau_total = wau.mean()
dau_total = dau.mean()

print('MAU: {:.0f}'.format(mau_total[0]))
print('WAU: {:.0f}'.format(wau_total[0]))
print('DAU: {:.0f}'.format(dau_total[0]))
```

### Рассчитаем Retention Rate, применяя когортный анализ
[К оглавлению](#21-bullet)

```python
#сгруппируем таблицу визитов по uid присваивая каждому uid первую дату визита
first_activity_date = visits.groupby(['uid'])['session_date'].min()
first_activity_date.name = 'first_activity_date'

#приджойним дату первого визита к таблице с визитами по uid, изменим тип данных,
#добавим необходимые столбцы с месяцами и рассчитаем время "жизни" когорт
user_activity = visits.join(first_activity_date,on='uid')
user_activity['first_activity_date'] = pd.to_datetime(user_activity['first_activity_date'])
user_activity['first_activity_month'] = user_activity['first_activity_date'].astype('datetime64[M]')
user_activity['activity_month'] = user_activity['session_date'].astype('datetime64[M]')
user_activity['cohort_lifetime'] = user_activity['activity_month'] - user_activity['first_activity_month']
user_activity['cohort_lifetime'] = user_activity['cohort_lifetime'] / np.timedelta64(1,'M')
user_activity['cohort_lifetime'] = user_activity['cohort_lifetime'].round().astype('int') 

#выделим в отдельную таблицу данные по когортам, их возрасту и количеству уникальных пользователей
cohorts = user_activity.groupby(['first_activity_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()

#выделим в отдельную таблицу данные о нулевой нулевом месяце "жизни" когорты
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_month','uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'}) 

#добавим данные о нулевой когорте к таблице cohorts и рассчитаем retention
cohorts = cohorts.merge(initial_users_count,on='first_activity_month')
cohorts['retention'] = cohorts['uid'] / cohorts['cohort_users'] 

#построим сводную таблицу по retention в разрезе когорт и их возруста 
retention_pivot = (cohorts.
                   pivot_table(index='first_activity_month',
                               columns='cohort_lifetime',
                               values='retention',
                               aggfunc='sum')
                  .fillna(0))
retention_pivot
```

```python
#визуализируем данные с помощью тепловой карты
sns.set(style='white')
plt.figure(figsize=(15, 7))
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=.05, linecolor='gray', vmax=.1, cmap="YlGnBu")
plt.title('Удержание пользователей по когортам (Retention Rate)')
plt.xlabel('Возраст когорты')
plt.ylabel('Когорты')
plt.show()
```

```python
#рассчитаем средний retention на 2 месяц "жизни" когорт
print('Средний retention на второй месяц жизни когорт {:.2%}'.format(retention_pivot.reset_index()[2][0:10].mean()))
```

В целом retention стабилен, колеблется в пределах от 2 до 10%, присутствуют выбросы. Например, интересно чем обусловлен высокий retention на второй месяц жизни когорты "2017-09-01" и как этот опыт можно распространить на остальные периоды и когорты. Возможно всплески были связаны с праздниками и предложением бонусов пользователям. Также есть провалы до 0 по некоторым когортам, с этим также необходимо разобраться и выяснять причины. Перечисленные вопросы стоит озвучить команде сервиса, возможно, на большинство из них у них есть ответы.

Средний retention на второй месяц жизни когорт 4.63%


## Метрики электронной коммерции
<a id='7-bullet'></a>
### Исследуем, сколько времени в среднем проходит с момента первого посещения до покупки
[К оглавлению](#21-bullet)

```python
#добавим данные о первом визите в таблицу с заказами, приведем типы данных к необходимым, добавим необходимые колонки
first_activity_uid = visits.groupby(['uid'])['start_ts'].min()
first_activity_uid.name = 'first_activity_date'
first_orders = orders.join(first_activity_uid, on='uid')
first_orders['first_activity_date'] = pd.to_datetime(first_orders['first_activity_date'])
first_orders['order_date'] = first_orders['buy_ts']
first_orders['days_to_first_order'] = first_orders['order_date'] - first_orders['first_activity_date']
first_orders['sec_to_first_order'] = first_orders['days_to_first_order'].dt.total_seconds()
first_orders.sample(7)
```

```python
#взглянем на распределение времени до первого заказа
first_orders['sec_to_first_order'].plot.hist(bins=150, figsize=(13, 8), color='black', range=(0, 7500))
plt.title('Распределение значений количества секунд, прошедших с момента первой активности до момента первой покупки')
plt.xlabel('Секунд до первого заказа')
plt.ylabel('Частота')
plt.show()
```

```python
#по графику видим, что большинство пользователей делало заказ в период не превышающий 4000 секунд,
#сгруппируем их и рассчитаем долю каждой
def group_to_order(sec_to_first_order):
    if sec_to_first_order <= 300:
        return '1 part (0-5 min)'
    if sec_to_first_order <= 600:
        return '2 part (5-10 min)'
    if sec_to_first_order <= 1200:
        return '3 part (10-20 min)'
    if sec_to_first_order <= 3600:
        return '4 part (20-60 min)'
    else:
        return '5 part (more 60 min)'

#сгруппируем данные по uid и найдем время до первой покупки для каждого
first_order_group = first_orders.groupby('uid').agg({'sec_to_first_order': 'min'})
first_order_group['group_to_order'] = first_order_group['sec_to_first_order'].apply(group_to_order)
first_order_group = first_order_group.reset_index()
first_order_group.columns = ['uid', 'sec_to_first_order', 'group_to_order']

first_order_pivot = first_order_group.pivot_table(index='group_to_order', values='sec_to_first_order', aggfunc='count')
first_order_pivot['part,%'] = ((first_order_pivot['sec_to_first_order'] / 
                              first_order_pivot['sec_to_first_order'].sum()) * 100).round(1)
first_order_pivot
```

```python
#большинство пользователей делало заказ в первый час, рассчитаем их долю
part_of_less_60min_to_order = (len(first_order_group.query('sec_to_first_order <= 3600')) / 
                             len(first_order_group['sec_to_first_order']))
print('Доля покупок в первый час после визита: {:.0%}'.format(part_of_less_60min_to_order))
```

Большинство пользователей (62%) сделало заказы в течении часа после визита.
При этом:
- 32% пользователей сделали заказ в первые 5 минут;
- 30% пользователей сделали заказ в период от 5 до 60 минут;
- 38% пользователей сделали заказ в период более 60 минут.

### Рассчитаем среднее количество покупок на одного клиента за период

```python
#добавим месяц покупки
orders['buy_ts_month'] = orders['buy_ts'].astype('datetime64[M]')

#сгруппируем таблицу заказов по uid присваивая каждому uid первый месяц заказа
first_orders_month = orders.groupby(['uid'])['buy_ts_month'].min()
first_orders_month.name = 'first_orders_month'

#приджойним дату первого заказа к таблице с заказами по uid, изменим тип данных
orders_cohorts = orders.join(first_orders_month, on='uid')
orders_cohorts['age'] = orders_cohorts['buy_ts_month'] - orders_cohorts['first_orders_month']
orders_cohorts['age'] = orders_cohorts['age'] / np.timedelta64(1,'M')
orders_cohorts['age'] = orders_cohorts['age'].round().astype('int')
print(orders_cohorts.duplicated().sum())

#сгруппируем данные по когортам и их возрасту, рассчитаем количество уникальных пользователей и количество заказов


orders_by_cohort = (orders_cohorts
                    .groupby(by=['first_orders_month', 'age'])
                    .agg({'uid': 'nunique'}))

uid_cohorts = orders_by_cohort.reset_index().query('age == 0')[['first_orders_month', 'uid']]
uid_cohorts.columns = ['first_orders_month', 'uid_cohort']

orders_new = (orders_cohorts
              .groupby(by=['first_orders_month', 'age'])
              .agg({'revenue': 'count'})
              .reset_index())
orders_new.columns = ['first_orders_month', 'age', 'count_orders']
orders_grouped = orders_new.merge(uid_cohorts, on='first_orders_month')

#orders_new = orders_by_cohort.merge(uid_cohorts, on='first_orders_month')
#orders_new
#рассчитаем среднее количество заказов на одного уникального пользователя
orders_grouped['buy_per_uid'] = orders_grouped['count_orders'] / orders_grouped['uid_cohort']
orders_grouped
#создадим сводную средних значений количества заказов в разрезе когорт и их возраста
orders_grouped_pivot = (orders_grouped
                        .pivot_table(index='first_orders_month',
                                     columns='age',
                                     values='buy_per_uid',
                                     aggfunc='mean')
                        .cumsum(axis=1)
                        .round(1))

#визуализируем данные с помощью тепловой карты
sns.set(style='white')
plt.figure(figsize=(15, 7))
sns.heatmap(orders_grouped_pivot, annot=True, fmt='.3', linewidths=.05, linecolor='gray', vmin=.5, cmap="YlGnBu")
plt.title('Среднее количество покупок на 1 пользователя в разрезе когорт и их "возраста" накопительным итогом')
plt.xlabel('"Возраст" когорт')
plt.ylabel('Когорты')
plt.show()
```

### Выясним, как меняется во времени накопительный LTV по когортам
[К оглавлению](#21-bullet)

```python
#добавим месяц первой покупки к каждому uid в таблице с заказами
first_orders = orders.groupby('uid').agg({'buy_ts_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']

#сгруппируем данные по когортам и рассчитаем количество уникальных пользователей на каждой из них
cohort_sizes = first_orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
cohort_sizes.columns = ['first_order_month', 'n_buyers']

#сгруппируем данные и рассчитаем возраст когорт, выручку каждой когорты по месяцам, ltv
orders_for_ltv = pd.merge(orders, first_orders, on='uid')
cohorts = orders_for_ltv.groupby(['first_order_month','buy_ts_month']).agg({'revenue': 'sum'}).reset_index()
report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
report['cohort_lifetime'] = ((report['buy_ts_month'] - report['first_order_month']) / 
                             np.timedelta64(1, 'M')).round().astype('int')       
report['ltv'] = report.revenue / report.n_buyers

#построим сводную таблицу для расчета накопительного ltv
ltv = (report
       .pivot_table(index='first_order_month', columns='cohort_lifetime', values='ltv',aggfunc='sum')
       .cumsum(axis=1)
       .round(2))

ltv.fillna('')
```

```python
#найдем первый визит и месяц первой покупки
users = visits.sort_values('start_ts').groupby('uid').first()
client = pd.merge(orders, users[['source_id']], on='uid', how='left')
client['first_order_ts'] = client.groupby('uid')['buy_ts'].transform('min')
client['first_order_month'] = client['first_order_ts'].astype('datetime64[M]')

#рассчитаем количество уникальных uid в разрезе когорт и источника
cohort_source_sizes = (client
                       .groupby(['first_order_month', 'source_id'])
                       .agg({'uid': 'nunique'})
                       .reset_index())
cohort_source_sizes.columns = ['first_order_month', 'source_id', 'n_buyers']

#рассчитаем сумму выручки в разрезе когорт, месяца покупки и источника
source_cohorts = (client
                  .groupby(['first_order_month','buy_ts_month', 'source_id'])
                  .agg({'revenue': 'sum'})
                  .reset_index())

#соединим таблицы, рассчитаем возраст когорт и ltv
source_report = pd.merge(cohort_source_sizes, source_cohorts, on=['first_order_month', 'source_id'])
source_report['cohort_lifetime'] = ((source_report['buy_ts_month'] - source_report['first_order_month']) / 
                                       np.timedelta64(1, 'M')).round().astype('int')
source_report['ltv'] = source_report['revenue'] / source_report['n_buyers']

#отфильтруем только когорты, которые "прожили" не менее 6 месяцев
cohort_6_month = ['2017-06-01', '2017-07-01', '2017-08-01', '2017-09-01', '2017-10-01', '2017-11-01', '2017-12-01']
source_report_6month = source_report.query('first_order_month in @cohort_6_month')

#создадим сводную таблицу накопительного ltv по когортам, источнику и возрасту когорты
source_ltv = (source_report_6month
              .pivot_table(index=['source_id', 'first_order_month'],
                           columns='cohort_lifetime', 
                           values='ltv',
                           aggfunc='sum')
              .cumsum(axis=1))
#рассчитаем средний накопительный ltv в разрезе когорт и источника за 6 месяцев
source_ltv_mean = source_ltv.round(2).fillna(0).reset_index().groupby('source_id').agg({5: 'mean'}).reset_index()
source_ltv_mean.columns = ['source_id', 'ltv6month']
source_ltv_mean
```

## Определим источники трафика и тип устройств, на которые маркетологам стоит делать упор
[К оглавлению](#21-bullet)

1. За 6 месяцев жизни окупились источники под номером 1 и 5, это самые перспективные источники трафика.
2. Вторыми по значимости идет пара источников под номерами 2 и 9, расходы на эти источники за 6 месяцев также практически окупились.
3. Источник под номером 4 спорный, за 6 месяцев окупился на 67%, однако за весь срок анализа уже на 93%.
4. Источник под номерм 10 за 6 месяцев окупился на 45%.
5. Источник под номером 3 окупился за 6 месяцев только на 30%, а за весь срок на 39%. Этот источник имеет самые большие расходы.
6. За весь срок анализа окупились источники 1, 2, 9 и 5, также близок к этому 4 источник. Лучше всего окупается источник под номером 1, хуже всего - 3. Источник под номером 10 - спорный, окупился почти на 80% за весь срок.
7. В основном на сайт заходят с десктопов, их доля в среднем составляет 71%. Вид девайсов "touch" имеет большой потенциал для роста, возможно, именно в этой группе стоит сосредоточить усилия.

## Суммируем выводы, которые мы сделали после подсчёта метрик каждого вида
[К оглавлению](#21-bullet)

1. В конце ноября можно наблюдать заметный рост пользовательской активности, далее следует спад. Предположим, что периоды пиковой активности связаны с выходными и праздничными/предпраздничными днями. Средние показатели активности: 23 228 уникальных пользователей в месяц; 5 716 уникальных пользователей в неделю; 908 уникальных пользователей в день.
2. За месяц на каждого уникального пользователя в среднем приходится не более 1,21 сессии, медиана количества сессий на уровне 1,08 сессии.
3. Средняя продолжительность сессии 644 секунды, но распределение не похоже на "нормальное" поэтому опираться на это значение нельзя, оно искажено выбросами. Рассчитав моду выборки продолжительности сессий можем сказать, что чаще всего пользователи проводят на сайте 60 секунд.
4. В целом retention стабилен, колеблется в пределах от 2 до 10%, присутствуют выбросы. Например, интересно чем обусловлен высокий retention на второй месяц жизни когорты "2017-09-01" и как этот опыт можно распространить на остальные периоды и когорты. Возможно всплески были связаны с праздниками и предложением бонусов пользователям. Также есть провалы до 0 по некоторым когортам, с этим также необходимо разобраться и выяснять причины. Перечисленные вопросы стоит озвучить команде сервиса, возможно, на большинство из них у них есть ответы. Средний retention на второй месяц жизни когорт 4.63%
5. Большинство пользователей (62%) сделало заказы в течении часа после визита. При этом 32% пользователей сделали заказ в первые 5 минут, 30% пользователей сделали заказ в период от 5 до 60 минут, 38% пользователей сделали заказ в период более 60 минут.
6. В среднем за 6 месяцев пользователи совершают по 1,44 заказа. Больше всего покупок совершили пользователи когорты "2017-06". Достаточно большой объем покупок был сделан в двух когортах ("2017-08" и "2017-11"). Необходимо проанализировать какие факторы повлияли на разницу среднего количества покупок в разных когортах.
7. Средний чек на уровне 5 у.е. В декабре вырос почти до 6, а в январе упал почти до 4. В остальное время показатель держится в коридоре 4.5 - 5.5, кроме июня. В июне была только одна покупка, поэтому результат этого месяца можно не учитывать.
8. Наибольшие значения LTV, практически на 50% превышающие показатели в прочих когортах по сопоставимым месяцам "жизни" наблюдаем в когорте 2017-09, необходимо проанализировать факторы, которые позволили достичь таких результатов и распространить, если возможно, это опыт на все прочие когорты. Средний LTV за 6 месяцев 7.97, а в когорте 2017-09 он на уровне 12.31.
9. Больше всего потратили на источник №3 в период с октября 2017 по март 2018 года. Также значительно потратились на источники 4, 5 и 2, на остальные источники траты были несущественные. В целом наибольшие расходы пришлись на период с октября по декабрь включительно. С декабря расходы на маркетинг падают.
10. Высокую стоимость привлечения из источника под номером 3 (12 уе) и 2 (10 уе) компенсировали стоимостью привлечения из других источников. Средняя стоимость привлечения: 8 уе. В разрезе источников самые дорогие источники привлечения под номерами 3 и 2 в период с июля по ноябрь включительно. В целом самая высокая стоимость привлечения приходится на август.
