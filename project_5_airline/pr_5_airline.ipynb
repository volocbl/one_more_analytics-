### 2. Проверим типы данных на корректность
[К оглавлению](#8-bullet)

```python
#изучим типы данных и наличие пропусков в таблице model_fl
model_fl.info()
```

```python
#изучим типы данных и наличие пропусков в таблице city_fl
city_fl.info()
```

```python
#приведем к целочисленным значениям столбец average_flights таблицы city_fl для удобства 
city_fl['average_flights'] = city_fl['average_flights'].astype('int')
city_fl.info()
```

```python
#проверим таблицу city_fl на наличие дубликатов
if city_fl.duplicated().sum() == 0:
    print('Полных дубликатов в таблице city_fl не обнаружено')
else:
    print('Есть полные дубликаты в таблице city_fl')

#проверим не задублировались ли города в таблице city_fl
if city_fl['city'].count() == len(city_fl):
    print('Дубликатов в столбце "city" таблицы city_fl также не обнаружено')
else:
    print('Обнаружены дубликаты в столбце "city" таблицы city_fl')
```

```python
#проверим таблицу model_fl на наличие дубликатов
if model_fl.duplicated().sum() == 0:
    print('Полных дубликатов в таблице model_fl не обнаружено')
else:
    print('Есть полные дубликаты в таблице model_fl')
    
#проверим не задублировались ли модели в таблице model_fl
if model_fl['model'].count() == len(model_fl):
    print('Дубликатов в столбце "model" таблицы model_fl также не обнаружено')
else:
    print('Обнаружены дубликаты в столбце "model" таблицы model_fl')
```

Отбросим значения после точки для столбца average_flights в таблице city_fl, чтобы не отвлекало от сути. Дубликатов в таблицах city_fl и model_fl не обнуружено.

### 3. Определим топ-10 городов по количеству рейсов
[К оглавлению](#8-bullet)

```python
#отсортируем по убыванию города по среднему числу рейсов и возьмем первые 10
city_fl.sort_values(by='average_flights', ascending=False).head(10)
```

Ожидаемо в первой тройке Москва, Санкт-Петербург и Новосибирск. При этом количество рейсов в Москву больше, чем во все остальные города из ТОП 10 вместе взятые.

<a id='4-bullet'></a>
### 4. Проанализируем модели самолетов и количество рейсов
[К оглавлению](#8-bullet)

```python
#взглянем на таблицу в моделями и количеством рейсов
model_fl
```

```python
#построим график количества рейсов по моделям самолетов
model_fl.sort_values(by='flights_amount', ascending=True)\
.plot(kind='barh',figsize=(8, 6), color='g', grid=True, x='model', y='flights_amount',
      title='Количество рейсов по моделям самолетов')
plt.xlabel('Количество рейсов')
plt.ylabel('Модель самолета')
plt.show()
```

Можно пердположить, что распределение объема рейсов по моделям связано с количеством мест на борту, так Cessna, в которой 14 мест совершает больше рейсов, чем Airbus A321-200 с пассожировместимостью от 170 до 220 человек. На последнем месте по объему рейсов Boeing 777-300 с пассажировместимостью в 550 человек.

<a id='5-bullet'></a>
### 5. Проанализируем города и количество рейсов
[К оглавлению](#8-bullet)

```python
#построим график среднего количества рейсов в разрезе городов
city_fl\
.sort_values(by='average_flights')\
.plot(kind='barh', figsize=(8, 25), color='b', grid=True, x='city', y='average_flights',
      title='Города и количество рейсов')
plt.xlabel('Среднее количетво рейсов')
plt.ylabel('Город')
plt.show()
```

Наиболее нагружены рейсами: Москва, Санкт-Петербург, Новосибирск, Красноярск и Екатеринбург. Наименее нагружены Ярославь, Псков, Нефтеюганск, Стрежевой и Петропавловловск-Камчатский. Можно предположить, что наиболее нагруженны рейсами центральные и  густонаселенные города.

<a id='6-bullet'></a>
### 6. Проанализируем топ-10 городов и количество рейсов
[К оглавлению](#8-bullet)

```python
#построим график топ 10 городов по среднему количеству рейсов
city_fl\
.sort_values(by='average_flights', ascending=False)\
.head(10)\
.sort_values(by='average_flights')\
.plot(kind='barh', figsize=(8, 6), color='r', grid=True, x='city', y='average_flights',
      title='ТОП 10 городов по количеству рейсов')
plt.xlabel('Среднее количество рейсов')
plt.ylabel('Город')
plt.show()
```

На первых местах Москва, Санкт-Пербург и Новосибирск. Эта картина в точности повторяет ТОП 3 по численности населения среди городов России. После Новосибирска наблюдается стабилизация и начинают влиять факторы, для оценки которых требуется больше данных.
